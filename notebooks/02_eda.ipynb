```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # data visualization
import seaborn as sns
import plotly.express as px # interactive visualizations

print("Libraries imported successfully!")
```

```python
df=pd.read_csv("../data/cleaned_customer_shopping_behavior.csv")
df.shape
df.isnull().sum()
```

```python
gendercounts=df['Gender'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gendercounts,labels=gendercounts.index,autopct='%1.1f%%',colors=['blue','pink'])
plt.title('Gender Distribution')
plt.show()
```

```python

plt.figure(figsize=(10,5))
sns.histplot(df['Age'], bins=20, color='blue', kde=True)
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Chart 3 - Top Product Categories
plt.figure(figsize=(10,5))
category_counts = df['Category'].value_counts()
print(category_counts)
sns.barplot(x=category_counts.index, 
            y=category_counts.values, 
            hue=category_counts.index,
            palette='viridis', 
            legend=False)
plt.title('Most Popular Product Categories')
plt.xlabel('Category')
plt.ylabel('Number of Purchases')
plt.show()
```

```python
# Chart 4 - Sales by Season
plt.figure(figsize=(8,5))
season_counts = df['Season'].value_counts()
sns.barplot(x=season_counts.index, 
            y=season_counts.values, 
            hue=season_counts.index,
            palette='coolwarm', 
            legend=False)
plt.title('Purchases by Season')
plt.xlabel('Season')
plt.ylabel('Number of Purchases')
plt.show()
```

```python
# Chart 5 - Payment Methods
plt.figure(figsize=(10,5))
payment_counts = df['Payment Method'].value_counts()
sns.barplot(x=payment_counts.index, 
            y=payment_counts.values, 
            hue=payment_counts.index,
            palette='magma', 
            legend=False)
plt.title('Preferred Payment Methods')
plt.xlabel('Payment Method')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

```python
# Chart 6 - Discount Impact on Purchase Amount
plt.figure(figsize=(8,5))
sns.boxplot(x='Discount Applied', y='Purchase Amount (USD)', 
            data=df, hue='Discount Applied', 
            palette='Set2', legend=False)
plt.title('Does Discount Affect Purchase Amount?')
plt.xlabel('Discount Applied')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

```python
# Chart 7 - Review Rating Distribution
plt.figure(figsize=(10,5))
sns.histplot(df['Review Rating'], bins=15, color='green', kde=True)
plt.title('Review Rating Distribution')
plt.xlabel('Review Rating')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Chart 8 - Purchase Amount Distribution
plt.figure(figsize=(10,5))
sns.histplot(df['Purchase Amount (USD)'], bins=20, color='orange', kde=True)
plt.title('Purchase Amount Distribution')
plt.xlabel('Purchase Amount (USD)')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Chart 9 - Subscription Status
plt.figure(figsize=(6,6))
sub_counts = df['Subscription Status'].value_counts()
plt.pie(sub_counts, labels=sub_counts.index, 
        autopct='%1.1f%%', colors=['lightgreen','lightcoral'])
plt.title('Customer Subscription Status')
plt.show()
```

```python
# Chart 10 - Frequency of Purchases
plt.figure(figsize=(12,6))
freq_counts = df['Frequency of Purchases'].value_counts()
sns.barplot(x=freq_counts.index, 
            y=freq_counts.values, 
            hue=freq_counts.index,
            palette='cubehelix', 
            legend=False)
plt.title('Frequency of Purchases')
plt.xlabel('Frequency')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```
